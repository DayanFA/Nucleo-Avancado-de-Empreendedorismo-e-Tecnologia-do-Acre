<h2 id="basic_grouping">4. Básico de Agrupamento usando método groupby</h2>


<p>O método "groupby" agrupa dados por diferentes categorias. Os dados são agrupados com base em uma ou várias variáveis, e a análise é realizada nos grupos individuais.</p>

<p>Por exemplo, vamos agrupar pela variável "drive-wheels". Vemos que existem 3 categorias diferentes de drive-wheels.</p>


````python
df['drive-wheels'].unique()
````

<p>Se quisermos saber, em média, qual tipo de drive wheels é mais valorizado em termos de preço, podemos agrupar "drive-wheels" e tirar a média deles.</p>

<p>Podemos selecionar as colunas 'drive-wheels', 'body-style' e 'price', em seguida atribua-o a variável "df_group_one".</p>


````python
df_group_one = df[['drive-wheels','body-style','price']]
````

Podemos então calcular o preço médio para cada uma das diferentes categorias de dados.


````python
# agrupamento com 1 variavel
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
````

<p>Olhando para a saída, parece que os veículos com tração traseira (rwd) são, em média, os mais caros, enquanto os de 4 rodas (4wd) e dianteiros (fwd) têm aproximadamente o mesmo price.</p>

<p>Você também pode agrupar por várias variáveis. Por exemplo, vamos agrupar por 'drive-wheels' e 'body-style'. Isso agrupa o dataframe pela combinação exclusiva de 'drive-wheeçs' e 'body-style'. Podemos armazenar os resultados na variável 'grouped_test1'. </p>

````python
# agrupamento com 2 variaveis
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
````

<p>Esses dados agrupados são muito mais fáceis de visualizar quando são transformados em uma tabela dinâmica. Uma tabela dinâmica é como uma planilha do Excel, com uma variável ao longo da coluna e outra ao longo da linha. Podemos converter o dataframe em uma tabela dinâmica usando o método "pivot" para criar uma tabela dinâmica a partir dos grupos.</p>

<p>Nesse caso, deixaremos a variável drive-wheels como as linhas da tabela e body-style para se tornar as colunas da tabela:</p>


````python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
````

<p>Frequentemente, não teremos dados para algumas das células dinâmicas. Podemos preencher essas células ausentes com o valor 0, mas qualquer outro valor também pode ser usado. Deve-se mencionar que a falta de dados é um assunto bastante complexo e é necessário um curso inteiro pra tratar deste assunto.</p>


````python
grouped_pivot = grouped_pivot.fillna(0) #preenche com 0 valores ausentes
grouped_pivot
````

<h2 id="correlation_causation">5. Correção e Causalidade</h2>


<p><b>Correlação</b>: uma medida do grau de interdependência entre duas variáveis.</p>

<p><b>Causalidade</b>: o relacionamento entre causa e efeito entre duas variáveis.</p>

<p>É importante saber a diferença entre elas duas. Correlação não implica em causalidade. Determinar correlação é muito mais simples que determinar causalidade, pois a causalidade pode exigir experimentação independente.</p>


<p><b>Correlação de Pearson</b></p>
<p>A correlação de Pearson mede a dependência linear entre duas variáveis X e Y.</p>
<p>O coeficiente resultante é um valor entre -1 e 1 inclusive, onde:</p>
<ul>
    <li><b>1</b>: Correlação linear positiva perfeita.</li>
    <li><b>0</b>: Sem correlação linear, as duas variáveis provavelmente não afetam uma à outra.</li>
    <li><b>-1</b>: Correlação linear negativa perfeita.</li>
</ul>


<p>Correlação de Pearson é o método padrão da função "corr". Como antes, calculamos a correlação de Pearson das variáveis 'int64' ou 'float64'.</p>


````python
df.corr()
````

É possível saber a significância da estimativa de correlação.


<b>P-value</b>

<p>O que é P-value? P-value é o valor de probabilidade de que a correlação entre essas duas variáveis seja estatisticamente significativa. Normalmente, escolhemos o nível de significância de 0.05, que significa que temos 95% de confiança de que a correlação entre as variáveis é significativa.</p>

Por conversão, quando

<ul>
    <li>o p-value é $<$ 0.001: dizemos que é forte a evidência que a correlação é significativa.</li>
    <li>o p-value é $<$ 0.05: há uma evidência moderada que a correlação é significativa.</li>
    <li>o p-value é $<$ 0.1: há uma fraca evidência que a correlação é significativa.</li>
    <li>o p-value é $>$ 0.1: não há evidência que a correlação é significativa.</li>
</ul>


Podemos obter essa informação usando o módulo "stats" da biblioteca "scipy".


````python
from scipy import stats
````

<h3>Wheel-Base vs. Price</h3>


Vamos calcular o *Coeficiente de Correação Pearson* e *P-value* de 'wheel-base' e 'price'.


````python
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
````

<h4>Conclusão:</h4>
<p>Como o p-value é $<$ 0.001, a correlação entre wheel-base e price é estatisticamente significativa, embora a relação linear não é extremamente forte (~0.585).</p>

Como o p-value é < 0.001, a correlação entre highway-mpg e price é estatisticamente significativa, e o coeficiente é cerca de -0.705 o que mostra que a relação linear é negativa e moderamente forte..
